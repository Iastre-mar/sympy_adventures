Данные для аппроксимации в таблице ниже:


```python
dct = {'масса':[1945, 1495, 1570, 1520], 'мощность двигателя':[560,340,343,431], 'время разгона':[4.3,4.9,5.2, None]}
```

```python
frame = pd.DataFrame.from_dict(dct)
frame
```

```python
a2,a1,a0 = symbols('a2,a1,a0')
```

```python
f1 = frame.iloc[0][0] * a2 + frame.iloc[0][1] * a1 + a0 - frame.iloc[0][2]
f1
```

```python
f2 = frame.iloc[1][0] * a2 + frame.iloc[1][1] * a1 + a0 - frame.iloc[1][2]
f2
```

```python
f3 = frame.iloc[2][0] * a2 + frame.iloc[2][1] * a1 + a0 - frame.iloc[2][2]
f3
```

```python
mse = 1/3 *((f1) ** 2 + (f2) ** 2 + (f3) ** 2)
mse
```

```python
mse_diff_a2 = diff(mse, a2)
mse_diff_a2
```

```python
mse_diff_a1 = diff(mse, a1)
mse_diff_a1
```

```python
mse_diff_a0 = diff(mse, a0)
mse_diff_a0
```
